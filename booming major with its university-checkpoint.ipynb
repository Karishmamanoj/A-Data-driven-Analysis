```python
import pandas as pd
```

```python
df=pd.read_excel('Dummydatanew.xlsx')
df
```

```python
df = df.rename(columns={'Name': 'School Name'})
```

```python
df.iloc[:, 2:] = df.iloc[:, 2:].abs()
df
```

```python
year_columns = [col for col in df.columns if col.isdigit()]

gradual_increase = df[df[year_columns].apply(lambda x: all(x[i] < x[i+1] for i in range(len(x)-1)), axis=1)]

gradual_increase
```

```python
filtered_df1 = gradual_increase[gradual_increase['2019'] != 0]

filtered_df1
```

```python
three_yrs_booming = filtered_df1.drop(['2018', '2019'], axis=1)

three_yrs_booming
```

```python
ratio_2020_2021 = (gradual_increase['2021'] - gradual_increase['2020']) / gradual_increase['2021'] * 100

ratio_2021_2022 = (gradual_increase['2022'] - gradual_increase['2021']) / gradual_increase['2022'] * 100

three_yrs_booming['ratio_2020_2021'] = ratio_2020_2021
three_yrs_booming['ratio_2021_2022'] = ratio_2021_2022

three_yrs_booming
```

```python
three_yrs_booming['Average_ratio'] = three_yrs_booming[['ratio_2020_2021', 'ratio_2021_2022']].mean(axis=1)

columns_to_round = ['ratio_2020_2021', 'ratio_2021_2022', 'Average_ratio']
three_yrs_booming[columns_to_round] = three_yrs_booming[columns_to_round].round(1)

three_yrs_booming[columns_to_round] = three_yrs_booming[columns_to_round].abs()

three_yrs_booming
```

```python
three_yrs_booming['Std_Dev'] = three_yrs_booming[['2020', '2021', '2022']].std(axis=1)
three_yrs_booming['Std_Dev'] = three_yrs_booming['Std_Dev'].round(1)
three_yrs_booming
```

```python
new_three_yrs_booming = three_yrs_booming.drop(['ratio_2020_2021', 'ratio_2021_2022'], axis=1)
new_three_yrs_booming
```

```python
new_three_yrs_booming = new_three_yrs_booming.rename(columns={'Name': 'School Name'})
```

```python
new_three_yrs_booming.to_excel('new_three_yrs_booming.xlsx',index=False)
```

```python
threshold_avg_int = 40
threshold_std_dev = 55

booming_courses_avg_int = new_three_yrs_booming[new_three_yrs_booming['Average_ratio'] >= threshold_avg_int][['School Name', 'Major', '2020', '2021', '2022', 'Average_ratio', 'Std_Dev']]
booming_courses_std_dev = new_three_yrs_booming[new_three_yrs_booming['Std_Dev'] > threshold_std_dev][['School Name', 'Major', '2020', '2021', '2022', 'Average_ratio', 'Std_Dev']]

booming_courses = pd.concat([booming_courses_avg_int, booming_courses_std_dev], ignore_index=True)
booming_courses = booming_courses[booming_courses['Average_ratio'] >= threshold_avg_int]
booming_courses = booming_courses[booming_courses['Std_Dev'] >= threshold_std_dev]

booming_courses
```

```python
sorted_in_desc = booming_courses.sort_values(by=['Std_Dev'], ascending=False)
sorted_in_desc
```

```python
sorted_in_desc_new=sorted_in_desc.drop_duplicates()
sorted_in_desc_new
```

```python
booming_majors_uni_yearwise = pd.merge(sorted_in_desc_new, filtered_df1[['School Name', 'Major', '2018', '2019']], on=['School Name', 'Major'], how='left')

booming_majors_uni_yearwise = booming_majors_uni_yearwise.rename(columns={'2020': '2020', '2021': '2021', '2022': '2022', 'Average_ratio': 'Average_ratio', 'Std_Dev': 'Std_Dev'})

column_order = ['School Name', 'Major', '2018', '2019', '2020', '2021', '2022', 'Average_ratio', 'Std_Dev']
booming_majors_uni_yearwise = booming_majors_uni_yearwise[column_order]

booming_majors_uni_yearwise
```

```python

booming_majors_uni_yearwise.to_excel('booming_majors_uni_yearwise.xlsx', index=False)
```

```python
rows = []

for index, row in booming_majors_uni_yearwise.iterrows():
    school_name = row['School Name']
    major = row['Major']
    for year in ['2018', '2019', '2020', '2021', '2022']:
        academic_year = f"{int(year)}/{int(year)+1}"
        grand_total = row[year]
        rows.append([school_name, major, academic_year, grand_total])

reshaped_booming_majors_uni_yearwise = pd.DataFrame(rows, columns=['School Name', 'Major', 'Academic Year', 'Grand Total'])

reshaped_booming_majors_uni_yearwise['Academic Year'] = reshaped_booming_majors_uni_yearwise['Academic Year'].apply(lambda x: x.split('/')[0])

reshaped_booming_majors_uni_yearwise
```

```python
reshaped_booming_majors_uni_yearwise.to_excel('reshaped_booming_majors_uni_yearwise.xlsx',index=False)
```
